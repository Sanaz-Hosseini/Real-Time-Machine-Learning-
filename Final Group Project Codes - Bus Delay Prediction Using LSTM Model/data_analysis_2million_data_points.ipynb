```python
#add a column for days based on date
day_of_week = df['service_date'].dt.day_name()
df.loc[:, 'day_of_week'] = day_of_week
#df.loc[:, 'day_of_week'] = df['service_date'].dt.day_name()
df
```

```python
#convert days to numbers
df['weekday'] = df['service_date'].dt.day_of_week
df
```

```python
#add a column (day_type) to keep the type of day (workday or weekend)
df['day_type'] = df['service_date'].apply(lambda x: 'Weekend' if x.weekday() >= 5 else 'Weekday')
df
```

```python
df['day_type']=df.day_type.factorize()[0]+ 1
df
```

```python
number_of_day_type_status=df.day_type.unique().size
print("number_of_day_types:"+ str(number_of_day_type_status))
```

```python
#convert scheduled_time to the format of hour:minute:second for all rows
df['scheduled_time']=pd.to_datetime(df['scheduled'], errors='coerce')
tindex=pd.DatetimeIndex(df.scheduled_time)
df['scheduled_time']=tindex.strftime('%H:%M:%S')
df
```

```python
#convert scheduled time to seconds based on comparison with to midnight
df['converted_scheduled_time']=pd.to_timedelta(df['scheduled_time']).dt.total_seconds()
df
```

```python
df['actual_time']=pd.to_datetime(df['actual'], errors='coerce')
tindex=pd.DatetimeIndex(df.actual_time)
df['actual_time']=tindex.strftime('%H:%M:%S')
df
```

```python
#Assume rush hour time in Boston occurs between 7:00-9:00 AM in the morning and 4:00-6:00 PM in the evening
conditions = [
    (new_df['converted_actual_time'] > 25200) & (new_df['converted_actual_time'] <= 32400),
    (new_df['converted_actual_time'] > 57600) & (new_df['converted_actual_time'] <= 64800)
    ]
values = [1,1]
new_df['rush_hour'] = np.select(conditions, values)
new_df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Group the dataframe by service_date, route_id, and half_trip_id
groups = new_df.groupby(['service_date', 'route_id', 'half_trip_id'])

rms_values = []
# Loop over the groups and extract the data for each group
for name, group in groups:
  rms_value = np.sqrt(np.mean(np.square(group['time_difference'])))
  rms_values.append(rms_value)
  #print(f"Group {name}: RMS value = {rms_value:.2f}")

mean_rms = np.mean(rms_values)
num_groups = groups.ngroups

print("Number of sequences:", num_groups)
print("Total Average of Mismatch time for all Sequences :", mean_rms)
#print("Number of rms:", len(rms_value))
plt.bar(range(len(rms_values)), rms_values, color='red' )
plt.xlabel('Sequence')
plt.ylabel('Average Mismatch Time (second)')
plt.title('Average Mismatch Time for Each Sequence')
plt.show()

first_group_name, first_group_data = list(groups)[0]
first_group_rms = np.sqrt(np.mean(np.square(first_group_data['time_difference'])))
print(f"First group name: {first_group_name}")
print(f"First group RMS value: {first_group_rms:.2f}")
for index, row in first_group_data.iterrows():
    print(row)
```

```python
import pandas as pd

# Group the dataframe by service_date, route_id, and half_trip_id
groups = new_df.groupby(['service_date', 'route_id', 'half_trip_id'])

rms_values = []
num_groups_lenght_9=0
# Loop over the groups and extract the data for each group
for name, group in groups:
  if len(group) == 9:
    rms_value = np.sqrt(np.mean(np.square(group['time_difference'])))
    rms_values.append(rms_value)
    num_groups_lenght_9 += 1
    #print(f"Group {name}: RMS value = {rms_value:.2f}")

mean_rms = np.mean(rms_values)
num_groups = groups.ngroups

print("Number of total sequences:", num_groups)
print("Number of sequences with lenght 9:", num_groups_lenght_9)
print("Average Mismatch:", mean_rms)
#print("Number of rms:", len(rms_value))
plt.bar(range(len(rms_values)), rms_values, color='red' )
plt.xlabel('Sequence')
plt.ylabel('Average Mismatch Time (second)')
plt.title('Average Mismatch time for Each Sequence with Lenght 9')
plt.show()

first_group_name, first_group_data = list(groups)[0]
first_group_rms = np.sqrt(np.mean(np.square(first_group_data['time_difference'])))
print(f"First group name: {first_group_name}")
print(f"First group RMS value: {first_group_rms:.2f}")
for index, row in first_group_data.iterrows():
    print(row)
```
